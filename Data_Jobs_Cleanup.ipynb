```python
#Importing libraries and tools
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Final output data to be used for vizualization
output_data_file = "output/data.csv"
```

```python
#Reading csv files
analyst_df = pd.read_csv("Data_Analyst.csv",encoding = "ISO-8859-1")
engineer_df = pd.read_csv("Data_Engineer.csv",encoding = "ISO-8859-1")
scientist_df = pd.read_csv("DataScientist.csv")
```

```python
#Sample cleanup for one of our datasets (data scientist)
df = scientist_df["Salary Estimate"].str.split("(", n = 1, expand = True) 
scientist_df["Salary Estimate"]= df
```

```python
#Splitting salary estimate column to min and max
df_1 = scientist_df["Salary Estimate"].str.split("-", n = 1, expand = True)
scientist_df["Salary Estimate Min"]= df_1[0]
scientist_df["Salary Estimate Max"]= df_1[1]

df_2 = scientist_df["Salary Estimate Min"].str.split("$", n = 1, expand = True)
scientist_df["Salary Estimate Min"]= df_2[1]

df_3 = scientist_df["Salary Estimate Max"].str.split("$", n = 1, expand = True)
scientist_df["Salary Estimate Max"]= df_3[1]
```

```python
#Converting units and data typesof salary column
scientist_df['Salary Estimate Max'] =  scientist_df['Salary Estimate Max'].str.replace('K','000',regex = False)
scientist_df['Salary Estimate Min'] =  scientist_df['Salary Estimate Min'].str.replace('K','000',regex = False)

#Replacing -1s with 0s
scientist_df.replace({"-1": "0"}, inplace=True)

#Dropping rows with outlier values
scientist_df.drop(scientist_df.index[scientist_df['Salary Estimate Max'] == '53 Per Hour'], inplace = True)
scientist_df.drop(scientist_df.index[scientist_df['Salary Estimate Max'] == '27 Per Hour'], inplace = True)
scientist_df.drop(scientist_df.index[scientist_df['Salary Estimate Max'] == '26 Per Hour'], inplace = True)

#Converting to integer
scientist_df['Salary Estimate Max'] = scientist_df['Salary Estimate Max'].astype(int)
scientist_df['Salary Estimate Min'] = scientist_df['Salary Estimate Min'].astype(int)
```

```python
#Removing unwanted characters from column
split_string = scientist_df["Company Name"].str.split("\n", n = 1, expand = True) 
scientist_df["Company Name"] =split_string[0]
scientist_df.head()
```

```python
#Parsing the job title column to extract the needed job titles for analysis
scientist_df.loc[scientist_df['Job Title'].str.contains('Data Analyst'), 'Job Title'] = "Data Analyst"
scientist_df.loc[scientist_df['Job Title'].str.contains('Scientist'), 'Job Title'] = "Data Scientist"
scientist_df.loc[scientist_df['Job Title'].str.contains('Engineer'), 'Job Title'] = "Data Engineer"
```

```python
#Replacing values
scientist_df['Job Title'] = scientist_df['Job Title'].replace({'Self Employed': 'Self-Employed', 'Self': 'Self-Employed'})
```

```python
#Using keywords we consolidated all jobs that had to do with data science into one universal name
scientist_df.loc[scientist_df['Job Title'].str.contains('Science'), 'Job Title'] = "Data Scientist"
scientist_df.loc[scientist_df['Job Title'].str.contains('SCIENCE'), 'Job Title'] = "Data Scientist  "
scientist_df.loc[scientist_df['Job Title'].str.contains('SCIENTIST'), 'Job Title'] = "Data Scientist"
scientist_df.loc[scientist_df['Job Title'].str.contains('science'), 'Job Title'] = "Data Scientist"
scientist_df.loc[scientist_df['Job Title'].str.contains('scientist'), 'Job Title'] = "Data Scientist"

#Only show Data Scientists
scientist_df = scientist_df.loc[scientist_df['Job Title'] == 'Data Scientist',:]
```

```python
#Creating a new dataframe with selected columns
scientist_df = scientist_df[['Job Title','Company Name','Rating','Salary Estimate Min','Salary Estimate Max','Location','Size','Founded','Type of ownership','Industry','Sector','Revenue','Easy Apply','Competitors']]
scientist_df['Salary Estimate Min'].astype(int)
scientist_df.head()
```

```python
#Preping the dataframes to be merged
analyst_df = analyst_df[['Job_Title','Company_Name','Rating','Minimum_Salary','Maximum_Salary','Location','Size','Founded','Ownership_Type','Industry','Sector','Revenue','Easy_Apply','Competitors']]

analyst_df = analyst_df.rename(columns = {'Job_Title': 'Job Title', 
                                          'Company_Name': 'Company Name',
                                          'Minimum_Salary': 'Salary Estimate Min', 
                                          'Maximum_Salary': 'Salary Estimate Max', 
                                        
                                          'Ownership_Type': 'Type of ownership', 
                                          'Easy_Apply': 'Easy Apply'}, inplace = False)
```

```python
#Preping the dataframes to be merged
engineer_df = engineer_df[['Job_Title','Company_Name','Rating','Minimum_Salary','Maximum_Salary','Location','Size','Founded','Ownership_Type','Industry','Sector','Revenue','Easy_Apply','Competitors']]
engineer_df = engineer_df.rename(columns = {'Job_Title': 'Job Title', 
                                          'Company_Name': 'Company Name',
                                          'Minimum_Salary': 'Salary Estimate Min', 
                                          'Maximum_Salary': 'Salary Estimate Max', 
                                          'Ownership_Type': 'Type of ownership', 
                                          'Easy_Apply': 'Easy Apply'}, inplace = False)
```

```python
#Consolidating three dataframes into one dataframe
frames = [engineer_df, analyst_df,scientist_df]
data = pd.concat(frames)
data.head()
```

```python
#Double checking we got all of the -1s out
data.replace({-1: 0}, inplace=True)
```

```python
#More cleanup
data.loc[data['Revenue'].str.contains('Unknown / Non-Applicable'), 'Revenue'] = '0'
```

```python
#Splitting and formating the revenue column
df_7 = data["Revenue"].str.split("(", n = 1, expand = True)
data["Revenue"] = df_7[0]

data['Revenue'] =  data['Revenue'].str.replace(' million','000000',regex = False)
data['Revenue'] =  data['Revenue'].str.replace(' billion','000000000',regex = False)

df_8 = data["Revenue"].str.split("to", n = 1, expand = True)
data["Revenue"] = df_8[1]

df_9 = data["Revenue"].str.split("$", n = 1, expand = True)
data["Revenue"] = df_9[1]


data["Revenue"] = data["Revenue"].replace([None], "0")
data
```

```python
#Creating a new column with States
df_4 = data["Location"].str.split(",", n = 1, expand = True)
df_4
data['State'] = df_4[1]
```

```python
#Catagorizing the Size column by our defined market segments
data.loc[data['Size'].str.contains('10000+ employees'), 'Size'] = "Enterprise(5,000+ Employees)"
data.loc[data['Size'].str.contains('1 to 50'), 'Size'] = "VSB (<50 Employees)"
data.loc[data['Size'].str.contains('5001 to 10000'), 'Size'] = "Enterprise(5,000+ Employees)"
data.loc[data['Size'].str.contains('51 to 200'), 'Size'] = "SMB(51-500 Employees)"
data.loc[data['Size'].str.contains('10000+'), 'Size'] = "Enterprise(5,000+ Employees)"
data.loc[data['Size'].str.contains('501 to 1000'), 'Size'] = "Mid-Market(501-5,000 Employees)"
data.loc[data['Size'].str.contains('501 to 1000'), 'Size'] = "Mid-Market(501-5,000 Employees)"
data.loc[data['Size'].str.contains('501 to 1000'), 'Size'] = "Mid-Market(501-5,000 Employees)"
```

```python
#Creating a new column that calculates average salary
data['Salary Estimate Average'] = (data['Salary Estimate Max'] + data['Salary Estimate Min']) / 2
data
```

```python
#Converting revenue column to more usable data
data['Revenue'] = data['Revenue'].astype(float)
```

```python
data['Revenue (millions)'] = data['Revenue'] / 10000000
data
```

```python
#Final columns to work with
data = data[['Job Title','Company Name','Rating','Salary Estimate Average','Salary Estimate Min','Salary Estimate Max','Location','State','Size','Founded','Type of ownership','Industry','Sector','Revenue (millions)','Easy Apply','Competitors']]
data
```

```python
data.to_csv("output/data.csv", index = False)
```
